# Classifier


```python
# Gradient  Boosting Classifier

from sklearn.datasets import make_classification
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
x,y = make_classification(n_samples=1000,n_features=10, random_state=1, n_classes=2)


#split dataset
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
clf = GradientBoostingClassifier()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
# Evaluate the report
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
# Hyper parameter tuning
from sklearn.model_selection import GridSearchCV

# Define the parameters grid

param_grid = {
    'n_estimators' :[100, 200, 300] ,
    'learning_rate' :[0.01, 0.1, 0.05, 0.2] ,
    'max_depth' :[3, 4, 5],
    'subsample' : [0.8, 0.9, 1.0],
    'min_samples_split' :[2, 5, 10] ,
    'min_samples_leaf' :[1, 2, 4]
}
```

```python
gbc = GradientBoostingClassifier()
grid_search = GridSearchCV(estimator=gbc, param_grid=param_grid, cv=5, verbose=3, n_jobs=-1)
```

```python
grid_search
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
best_model = grid_search.best_estimator_
```

```python
y_pred_tuned =best_model.predict(X_test)
```

```python
# evaluate
print(accuracy_score(y_test, y_pred_tuned))
print(classification_report(y_test, y_pred_tuned))
print(confusion_matrix(y_test, y_pred_tuned))
```

# Regressor

```python
# Gradient  Boosting Regressor

from sklearn.datasets import make_regression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
x, y = make_regression(n_samples=1000, n_features=10, random_state=1, noise=10)

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
gbr = GradientBoostingRegressor()
gbr.fit(X_train, y_train)
```

```python
y_pred = gbr.predict(X_test)
```

```python
print(r2_score(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
```

```python
# Hyper Param

param_gridr = {
    'n_estimators' :[100, 200, 300] ,
    'learning_rate' :[0.01, 0.1, 0.05, 0.2]
}
```

```python

grid_searchr = GridSearchCV(estimator=gbr, param_grid=param_gridr, cv=5, verbose=3, n_jobs=-1)
```

```python
grid_searchr.fit(X_train, y_train)
```

```python
grid_searchr.best_params_
```

```python
best_modelr = grid_searchr.best_estimator_
```

```python
y_pred_tuned_r = best_modelr.predict(X_test)
```

```python
y_pred_tuned_r
```

```python
print(r2_score(y_test, y_pred_tuned_r))
print(mean_absolute_error(y_test, y_pred_tuned_r))
print(mean_squared_error(y_test, y_pred_tuned_r))
```
